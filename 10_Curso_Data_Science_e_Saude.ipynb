```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths)
display(recovered)
```

```python
deaths.groupby("Country/Region").sum()[last_date]
```

```python
recovered.groupby("Country/Region").sum()[last_date]
```

```python
def latest_by_country(data):
  # return data.groupby("Country/Region").sum()[last_date]
  return data.groupby("Country/Region").sum().iloc[:,-1] #pegando sempre a última coluna

latest_by_country(confirmed)
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
pd.concat(informations)
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
pd.concat(informations, axis=1)
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined
```

```python
combined.info()
```

```python
# taxa de letalidade em %
combined.sum()
```

```python
sum_up = combined.sum()
```

```python
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

```python
letality_rate_1 = combined["deaths"] / combined["confirmed"] *100
letality_rate_2 = combined["deaths"] / (combined["recovered"] + combined["deaths"]) *100
combined["letality_rate_1"] = letality_rate_1
combined["letality_rate_2"] = letality_rate_2

combined
```

```python
def latest_by_country_at (data, date):
  return data.groupby("Country/Region").sum()[date]
```

```python
informations_2 = [latest_by_country_at(confirmed, '2/28/23'), latest_by_country_at(deaths,'2/28/23'), latest_by_country_at(recovered,'2/28/23')]
combined_2 = pd.concat(informations_2, axis=1)
combined_2.columns = ['confirmed', 'deaths', 'recovered']
combined_2
```

```python
letality_rate_1 = combined_2["deaths"] / combined_2["confirmed"] *100
letality_rate_2 = combined_2["deaths"] / (combined_2["recovered"] + combined_2["deaths"]) *100
combined_2["letality_rate_1"] = letality_rate_1
combined_2["letality_rate_2"] = letality_rate_2

combined_2
```
