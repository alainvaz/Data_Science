# Exercício

Suponha que estejamos testando as moedas que serão utilizadas por árbitros de futebol nas competições da Copa do Mundo. Atualmente, estas moedas são personalizadas e utilizadas em certos momentos dos jogos para decidir, na sorte, uma disputa. Nosso objetivo é verificar se as moedas que serão utilizadas nas competições não sejam viciadas.

Para isso testes foram realizados com cinco moedas e os resultados foram os seguintes:

```python
m1 = 'CCcCCccCCCccCcCccCcCcCCCcCCcccCCcCcCcCcccCCcCcccCc'
m2 = 'CCCCCccCccCcCCCCccCccccCccCccCCcCccCcCcCCcCccCccCc'
m3 = 'CccCCccCcCCCCCCCCCCcccCccCCCCCCccCCCcccCCCcCCcccCC'
m4 = 'cCCccCCccCCccCCccccCcCcCcCcCcCcCCCCccccCCCcCCcCCCC'
m5 = 'CCCcCcCcCcCCCcCCcCcCCccCcCCcccCccCCcCcCcCcCcccccCc'
```

Acima, temos o resultado de 50 lançamentos de cada moeda (m1, m2, m3, m4 e m5), onde c representa a ocorrência do evento CARA e C representa a ocorrência do evento COROA.

Para tirarmos nossas conclusões, precisamos montar o seguinte DataFrame:


Faces	m1	m2	m3	m4	m5
C	Cara	Freq. C	Freq. C	Freq. C	Freq. C	Freq. C
c	Coroa	Freq. c	Freq. c	Freq. c	Freq. c	Freq. c

Onde Freq. C e Freq. c são, respectivamente, as frequências de COROAS e de CARAS em cada teste.

Gere um código para realizar tal tarefa.

```python
eventos = {'m1': list(m1),
            'm2': list(m2),
            'm3': list(m3),
            'm4': list(m4),
            'm5': list(m5)}
moedas = pd.DataFrame(eventos)
df = pd.DataFrame(data = ['Cara', 'Coroa'],
                    index = ['c', 'C'],
                    columns = ['Faces'])
for item in moedas:
    df = pd.concat([df, moedas[item].value_counts()],
                    axis = 1)
df
```

# Removendo outliers

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

Teremos uma representação gráfica que nos ajudará a compreender a técnica que usaremos nesta aula para remoção de outliers. Estamos utilizando Box-plot, que possui a seguinte configuração: temos uma mediana, em que dividimos os dados em 50%, para a direita e esquerda, igualmente. Teremos o Q1, que se refere ao primeiro quartil e parte em 25% e 75%, já o Q3 parte os dados em 75% e 25%. A diferente entre Q3 e Q1 gera o intervalo interquartirico, isto é, as estatísticas que geraremos para realizar o corte de outliers.

```python
dados_outliers = pd.read_csv('/content/aluguel_tipo_agregado.csv', sep=';')
```

```python
IIQ = Q3 - Q1
IIQ
```

```python
limite_inferior = Q1 - 1.5 * IIQ
limite_inferior
```

Identificando e Removendo Outliers

```python
dados_outliers.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados_outliers.groupby('Tipo')
```

```python
type(grupo_tipo)
```

```python
grupo_tipo_2 = dados_outliers.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo_2)
```

```python
grupo_tipo_2.groups
```

```python
Q1 = grupo_tipo_2.quantile(.25)
Q3 = grupo_tipo_2.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
Q1
```

```python
Q3
```

```python
IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
limite_superior['Apartamento']
```

```python
for tipo in grupo_tipo_2.groups.keys():
  print(tipo)
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo_2.groups.keys():
    eh_tipo = dados_outliers['Tipo'] == tipo
    eh_dentro_limite = (dados_outliers['Valor'] >= limite_inferior[tipo]) & (dados_outliers['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados_outliers[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```
